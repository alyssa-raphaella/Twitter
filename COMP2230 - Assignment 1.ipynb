```python
import pandas as pd
import os
import re

import warnings; warnings.simplefilter('ignore')
```

```python
# Initialize Twitter Dataframe (consolidated data)
twitter_df = pd.DataFrame()
# Get all data from 'data/*.csv', dump to twitter_df DataFrame
path = os.getcwd()
directory = os.path.join(path)
for root, dirs, files in os.walk(directory):
    for file in files:
        if file.endswith(".csv"):
            df_dump = pd.read_csv('data\\' + file)
            twitter_df = pd.concat([twitter_df, df_dump])
```

```python
twitter_df.columns = ['Timestamp', 'ID', 'Tweet']
twitter_df.drop_duplicates(subset = 'Tweet', keep = 'last', inplace = True)
twitter_df.Timestamp = pd.to_datetime(twitter_df['Timestamp'])
twitter_df = twitter_df.set_index('ID')
```

```python
length_df = len(twitter_df)
x = 0

def deEmojify(inputString):
    return inputString.encode('ascii', 'ignore').decode('ascii')

while x < length_df:
    content = twitter_df['Tweet'].iloc[x]
    content = deEmojify(content)
    content = re.sub(r"http\S+", "", content)
    content = content.replace('@ScottMorrisonMP', 'Scott Morrison')
    content = re.sub(r"@(\w+)", ' ', content, flags=re.MULTILINE)
    content = content.replace('#', ' ')
    twitter_df['Tweet'].iloc[x] = content.replace('\n', ' ')
    x = x + 1
    
twitter_df.drop_duplicates(subset = 'Tweet', keep = 'last', inplace = True)
```
